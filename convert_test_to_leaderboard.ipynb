# Convert test_set_day1.txt to leaderboard-style JSON
Reads questions, strips blank lines, assigns string IDs, writes leaderboard_quesies.json.

```python
from pathlib import Path
import json

input_path = Path("test_set_day1.txt")
output_path = Path("quesies.json")

questions = [line.strip() for line in input_path.read_text(encoding="utf-8").splitlines() if line.strip()]

data = [{"question": q, "id": str(i + 1)} for i, q in enumerate(questions)]

output_path.write_text(json.dumps(data, indent=4, ensure_ascii=True), encoding="utf-8")

print(f"Wrote {len(data)} questions to {output_path}")
```
